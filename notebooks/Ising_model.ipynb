Image Denoising
---------------

Let's assume we have a noisy, binary image, and we want to recover the original image. The code below will create our binary image coded as -1, 1 and the noisy version.

```python
from PIL import Image
import numpy as np
from scipy.signal import convolve2d
from scipy import stats
```

```python
img = imread('lettera.bmp')
arr = np.asarray(img, int)
arr = arr[:0:-1,:0:-1]  # Flip it
mn = np.mean(arr)
arr[arr < mn] = -1
arr[arr >= mn] = 1
original_mean = np.mean(arr)
print original_mean
```

```python
sigma = 2
noisy_arr = arr + sigma*np.random.normal(size=arr.shape)
```

```python
imshow(noisy_arr)
```

```python
noisy_mean = np.mean(noisy_arr)
print noisy_mean
imshow(noisy_arr > noisy_mean)
```

The Ising Model
---------------

How are we to recover the original image? The key insight is that *a pixel's value is usually correlated with its neighbors*. Thus, we cast the problem as a Markov Random Field. The probability of an image y is

$$ \log \tilde{p}(y) = - \sum_{s \tilde t } y_s w_{st} y_t$$

There are a few observations to be made here. First, the weight $ w_{st} $ describes the amount that the values of pixels $y_s$ and $y_t$ correspond to each other. If $w_{st} > 1$ then the probability is higher if $y_t$ and $y_s$ share the same value. This follows because $-1 * 1 * -1 = 1 * 1* 1 = 1$ but $1 * 1 * -1 = -1 * 1 * 1 = 0.$ It is often informative to think about Ising models in terms of a sum over products of pixel values encoded as -1, 1. Computationally, it's easier to treat them as a convolution, which we will see. Second, we assume that t and s, the pixel locations, are near each other in the image. If that isn't true, then the weight is 0. We can recast the equation as 

$$ \log \tilde{p}(y) = -\frac{1}{2} y^T W y $$

where W is a sparse block Toeplitz matrix. In keeping with our insight, we set $w_{st} = 1$ iff $y_s$ and $y_t$ are adjacent pixels. In fact, if we create a convolution matrix to simulate W, we see that the noise is somewhat reduced. 

```python
w_conv = np.ones((3, 3)) 
w_conv[1, 1] = 0
imshow(convolve2d(noisy_arr, w_conv, mode='same'))  # mode='same' to keep same size
```

Mean Field Update (Variational Inference)
----------------------------

Now, let's recover the original image. To do this, we'll need to set up a bit of notation. Call the noisy image $y$ and the output image $x$, both of which are vectorized images (though our code uses convolve2, so we keep them as actual images). We want to find

$$ min_x p(y, x) = p(x) p(y|x) $$

Our prior, $p(x)$ is constructed from the Ising model, so it encodes the idea that nearby pixels influence each other:

$$ p(x) = \frac{1}{Z_0} exp(-E_0(x)) = \frac{1}{Z_0} exp(-\sum_i\sum_{j \in nbhd(i)} W_{ij} x_i x_j) $$

Our likelihood is given by 

$$ p(y|x) = \prod_i p(y_i|x_i) = \sum_i exp(-L_i(x_i)) $$

for log likelihood loss function $L_i.$ That gives us a posterior:

\begin{equation}
 p(x|y) = \frac{1}{Z} exp(-E(X)) = \frac{1}{Z} exp(\sum_i L_i(x_i) - E_0(x)) 
\end{equation}

(Note: we don't need to know $Z$. Below, we use $\tilde p = pZ$

The prior energy is defined above, and it introduces a dependency between $x_i$ and its neighboring pixels in the image as described by the Ising matrix $W$. We seek a simple method to optimize the probability, but those dependencies introduce a lot of partial derivatives.

Instead, we'll use the mean field algorithm, which is a variation of Variational Inference. As a quick review, Variational Inference is an approximation technique for a probability distribution $p$. The idea is to create a class of distributions $q$ that are easier to work with, and then to parameterize $q$ to minimize the approximation error. 


In this case, our simplification is to assume that the posterior fully factorizes to 

$$ q(x) = \prod_i q_i(x_i, \mu_i) $$

where parameter $\mu_i$ is the mean value at pixel $i.$ Note that (1) we can choose $\mu_i$ to minimize $q$: it is a variational parameter, and (2) those means are independent of the surrounding pixels. That means we can derive the mean field update for each pixel independently. We start with 

$$ \log \tilde{p} (x_i) = x_i \sum_{j \in nbhd(i)} W_{ij} x_j + L_i(x_i) + const .$$

For the mean field update, we need to compute (see Murphy section 21.3.1 for details)

$$ \log q_j(x_j) = \mathbb E_{-q_j} [\log \tilde p(x)] + const $$

and since $E_{-q_j} (f) = \sum_{k \neq j} q(x_j, \mu_j | x_j) f(j) = \sum_{k \neq j} q(\mu_j) f(j), $ we have

$$ q_i(x_i) \propto exp(x_i \sum_{j \in nbhd(i)} W_{ij} \mu_j + L_i(x_i))  .$$

I don't want to copy Murphy's derivation of the actual update, which follows. Suffice to say, it uses $L_i^+ \equiv L_i(+1)$ and $L_i^- \equiv L_i(-1)$ which are the log likelihood functions centered at each of these two values. The variance in the likelihood controls the strength of the prior. We are left with this update, which also incorporates a damping term:

$$ \mu_i^t = (1 - \lambda)\mu_i + \lambda \tanh \big( x_i \sum_{j \in nbhd(i)} W_{ij} \mu_j + 0.5 (L_i^+  - L_i^-) \big) $$

Damping is required because the mean field Ising update alone can lead to clumping around local extrema.

It's actually pretty easy to compute this in python. What we do below is iterate through all of the pixels at once, using the convolution parameter as our mean function.


```python
noisy_arr_copy = noisy_arr.copy()
lmbda = 0.5
for i in range(15):
    logodds = log(stats.norm.pdf(noisy_arr_copy, loc=1, scale=2)) - log(stats.norm.pdf(noisy_arr_copy, loc=-1, scale=2))
    noisy_arr_copy = (1 - lmbda) * noisy_arr_copy + lmbda * tanh(convolve2d(logodds, w_conv, mode='same') + .5 * logodds)
```

```python
imshow(noisy_arr_copy)
print sum(np.abs(noisy_arr_copy - arr)) / arr.size
```

```python
denoised_mu = np.mean(noisy_arr_copy)
noisy_arr_copy[noisy_arr_copy < denoised_mu] = -1
noisy_arr_copy[noisy_arr_copy >= denoised_mu] = 1
```
